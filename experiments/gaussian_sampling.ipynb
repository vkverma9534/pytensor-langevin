```python
import numpy as np
import matplotlib.pyplot as plt

from models.banana import BananaModel
from samplers.ula import ULASampler
from samplers.mala import MALASampler
```

```python
model = BananaModel()
```

```python
ula = ULASampler(model, step_size=0.05)

ula_samples = ula.sample(x0=[0.0, 0.0], n_samples=8000)
```

```python
mala = MALASampler(model, step_size=0.2)

mala_samples, acc_rate = mala.sample(x0=[0.0, 0.0], n_samples=8000)

print("MALA acceptance rate:", acc_rate)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(ula_samples[:,0])
plt.title("ULA trace (x1)")

plt.subplot(1,2,2)
plt.plot(ula_samples[:,1])
plt.title("ULA trace (x2)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.scatter(
    ula_samples[:,0],
    ula_samples[:,1],
    s=5,
    alpha=0.3,
    label="ULA"
)

plt.scatter(
    mala_samples[:,0],
    mala_samples[:,1],
    s=5,
    alpha=0.3,
    label="MALA"
)

plt.legend()
plt.title("Samples from Banana Distribution")
plt.xlabel("x1")
plt.ylabel("x2")

plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(ula_samples[:,0], alpha=0.7)
plt.title("ULA trace x1")

plt.subplot(1,2,2)
plt.plot(mala_samples[:,0], alpha=0.7)
plt.title("MALA trace x1")

plt.tight_layout()
plt.show()
```
